```python
#text transcription
import whisper

model = whisper.load_model("medium")
result = model.transcribe("/content/Small Talk _ Everyday English.mp3", task="transcribe")
print(result["text"])
language = result['language']

```

```python
from transformers import MarianMTModel, MarianTokenizer
print('en = English\nfr = French\nes = Spanish\nde = German\nit = Italian\npt = Portuguese\nar = Arabic\nzh = Chinese (Simplified)\nja = Japanese\nko = Korean\nru = Russian\nhi = Hindi\nnl = Dutch\npl = Polish\ntr = Turkish\nsv = Swedish\nno = Norwegian\nfi = Finnish\nel = Greek\nth = Thai')
other_language = input('choose a language to translate into')
model_name = 'Helsinki-NLP/opus-mt-'
model_name += language
model_name += '-'
model_name += other_language
tokenizer = MarianTokenizer.from_pretrained(model_name)
model = MarianMTModel.from_pretrained(model_name)
```

```python
text = result["text"]
inputs = tokenizer(text, return_tensors="pt", padding=True,truncation=True)
translated = model.generate(**inputs)
print(tokenizer.decode(translated[0], skip_special_tokens=True))
```

```python
from gtts import gTTS
from IPython.display import Audio

text = tokenizer.decode(translated[0], skip_special_tokens=True)

tts = gTTS(text,lang = other_language)
tts.save("output.mp3")
Audio("output.mp3", autoplay=True)
```
